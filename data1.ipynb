```
data=pd.read_csv("datasets_674388_1186156_tv_shows.csv",index_col=None)
```

```
data.drop(columns='Unnamed: 0',inplace=True)
```

```
data.head(50)
```

```
data_long=pd.melt(data[['Title','Netflix','Hulu','Disney+','Prime Video']],id_vars=['Title'],var_name='Streaming Service', value_name='Present')
data_long = data_long[data_long['Present'] == 1]
data_long.drop(columns=['Present'],inplace=True)
```

```
data_combined = data_long.merge(data, on='Title', how='inner')
```

```
data_combined.drop(columns = ['Netflix','Hulu', 'Prime Video', 'Disney+', 'type'], inplace=True)
```

```
data_combined.head(5)
```

```
import plotly
import plotly.express as px
```

```
px.scatter(data_combined, x='IMDb',y='Rotten Tomatoes',color='Streaming Service')
```

```
data_age=data_combined.groupby(['Streaming Service',"Age",'Year'])[['Title']].count().reset_index()
```

```
data_age.head(5)
```

```
px.bar(data_age, x='Year',y='Age',color='Streaming Service')
```
